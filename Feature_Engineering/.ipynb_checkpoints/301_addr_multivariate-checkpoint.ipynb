```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import os, gc, datetime, time
pd.set_option("display.max_rows", 500)
pd.set_option("display.max_columns", 500)
from plotnine import *
import warnings
warnings.filterwarnings("ignore")
```

```python
train_idf = pd.read_pickle('/home/jeong/kaggle/IEEE/input/train_identity.pkl')
train_df = pd.read_pickle('/home/jeong/kaggle/IEEE/input/train_transaction.pkl')

test_idf =pd.read_pickle('/home/jeong/kaggle/IEEE/input/test_identity.pkl')
test_df = pd.read_pickle('/home/jeong/kaggle/IEEE/input/test_transaction.pkl')
```

```python
train_df['dist_log'] = np.log(train_df['dist2'] + 1 )
test_df['dist_log'] = np.log(test_df['dist2'] + 1 )

df = train_df[['TransactionID','dist_log']].append(test_df[['TransactionID','dist_log']])
df['dist_qcut'] = pd.qcut(df['dist_log'], 8)

df.drop('dist_log', axis = 1, inplace = True)

train_df = train_df.merge(df, on = 'TransactionID', how = 'left')
test_df = test_df.merge(df, on = 'TransactionID', how = 'left')

train_df['dist_qcut_addr1_addr2'] = train_df['dist_qcut'].astype('str') + train_df['addr1'].astype('str') + train_df['addr1'].astype('str')
test_df['dist_qcut_addr1_addr2'] = test_df['dist_qcut'].astype('str') + test_df['addr1'].astype('str') + test_df['addr1'].astype('str')

train_df['dist_qcut_addr1_addr2_Prod'] = train_df['dist_qcut'].astype('str') + train_df['addr1'].astype('str') + train_df['addr1'].astype('str') + train_df['ProductCD'].astype('str')
test_df['dist_qcut_addr1_addr2_Prod'] = test_df['dist_qcut'].astype('str') + test_df['addr1'].astype('str') + test_df['addr1'].astype('str') + test_df['ProductCD'].astype('str')

train_df['dist_qcut_addr1_addr2'] = train_df['dist_qcut_addr1_addr2'].astype('object')
test_df['dist_qcut_addr1_addr2']  = test_df['dist_qcut_addr1_addr2'].astype('object')

train_df['dist_qcut_addr1_addr2_Prod'] = train_df['dist_qcut_addr1_addr2_Prod'].astype('object')
test_df['dist_qcut_addr1_addr2_Prod'] = test_df['dist_qcut_addr1_addr2_Prod'].astype('object')

train_df.drop(['dist_log','dist_qcut'], axis = 1, inplace = True)
test_df.drop(['dist_log','dist_qcut'], axis = 1, inplace = True)
```

전달 대비 증감률

```python
import datetime 

start_date = datetime.datetime.strptime('2017.11.30', '%Y.%m.%d')
train_df['timeblock'] = train_df['TransactionDT'].apply(lambda x: datetime.timedelta(seconds = x) + start_date ) 
test_df['timeblock'] = test_df['TransactionDT'].apply(lambda x: datetime.timedelta(seconds = x) + start_date ) 


train_df['yearmonth'] = (100 * pd.to_datetime(train_df['timeblock']).dt.year) + (1 * pd.to_datetime(train_df['timeblock']).dt.month)
test_df['yearmonth'] = (100 * pd.to_datetime(test_df['timeblock']).dt.year) + (1 * pd.to_datetime(test_df['timeblock']).dt.month)

temp = train_df.groupby(['addr1']).size().nlargest(20).reset_index()
id2 = temp.addr1.values.tolist()

tr = train_df[train_df['addr1'].isin(id2)]
te = test_df[test_df['addr1'].isin(id2)]

df = tr[['yearmonth','addr1']].append(te[['yearmonth','addr1']])

temp = df.groupby(['yearmonth','addr1']).size().reset_index()
temp.columns = ['yearmonth','addr1','yearmonth_addr1_cnt']
temp.sort_values(['addr1','yearmonth'], inplace = True)

temp['yearmonth_addr1_cnt_lastmonth'] = temp.groupby('addr1')['yearmonth_addr1_cnt'].shift(1)

temp['yearmonth_addr1_ratio'] = temp['yearmonth_addr1_cnt'] / temp['yearmonth_addr1_cnt_lastmonth']
temp.head(20)

temp = temp[['yearmonth','addr1','yearmonth_addr1_ratio']]

train_df = train_df.merge(temp, on = ['yearmonth','addr1'], how = 'left' )
test_df = test_df.merge(temp, on = ['yearmonth','addr1'], how = 'left' )

train_df.drop(['yearmonth'], axis = 1, inplace = True)
test_df.drop(['yearmonth'], axis = 1, inplace = True)
```
